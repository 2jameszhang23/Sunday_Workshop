```python
#turtle.forward(distance)：向前移动指定的距离。
#turtle.backward(distance)：向后移动指定的距离。
#turtle.right(angle)：顺时针旋转指定的角度。
#turtle.left(angle)：逆时针旋转指定的角度。
#turtle.goto(x, y)：移动到指定的坐标 (x, y)。
#turtle.setheading(angle)：将海龟的方向设置为指定的角度（0为东，90为北，180为西，270为南）。
#turtle.penup()：抬起画笔，移动时不绘制线条。
#turtle.pendown()：放下画笔，开始绘制线条。

#turtle.circle(radius)：画一个指定半径的圆，默认从当前方向开始画圆。
#turtle.begin_fill()：开始填充颜色。
#turtle.end_fill()：结束填充颜色并封闭形状。

#turtle.color(colorname)：设置画笔和填充颜色（可以传入颜色名称，如 "red"，也可以传入 RGB 值）。
#turtle.fillcolor(colorname)：设置填充颜色。
#turtle.pencolor(colorname)：设置画笔颜色。

#turtle.width(width)：设置画笔的宽度。
#turtle.pensize(width)：与 width() 一样，设置画笔的宽度。
#turtle.speed(speed)：设置海龟的移动速度，speed 可以是 1-10 或者 "fastest", "fast", "normal", "slow", "slowest"。
#turtle.isdown()：检查画笔是否放下，返回 True 或 False。

#turtle.bgcolor(colorname)：设置背景颜色。
#turtle.title(title)：设置窗口的标题。
#turtle.setup(width, height)：设置绘图窗口的宽度和高度。
#turtle.clearscreen()：清除屏幕上的所有绘图。
#turtle.reset()：清除屏幕并将海龟重置为初始状态。
 
#turtle.position()：返回当前海龟的位置 (x, y)。
#turtle.heading()：返回当前海龟的方向（角度）。
#turtle.xcor()：返回当前海龟的 x 坐标。
#turtle.ycor()：返回当前海龟的 y 坐标。

#turtle.hideturtle()：隐藏海龟。
#turtle.showturtle()：显示海龟。
#turtle.clear()：清除当前海龟的绘图，但不改变海龟的位置和状态。
#turtle.bye()：关闭 turtle 图形窗口并退出程序。
```

```python
import turtle

# 设置画布分辨率和背景颜色
turtle.setup(width=500, height=500)  # 设置画布大小为500x500像素
turtle.bgcolor("lightblue")  # 设置背景颜色

# 创建海龟对象
t = turtle.Turtle()
t.speed(3)

# 画笑脸的脸部（大圆）
t.penup()
t.goto(0, -200)  # 将海龟移动到适当位置开始画圆
t.pendown()
t.color("yellow")
t.begin_fill()
t.circle(200)  # 画一个半径为200的圆
t.end_fill()

# 画笑脸的左眼
t.penup()
t.goto(-70, 50)  # 移动到左眼的位置
t.pendown()
t.color("black")
t.begin_fill()
t.circle(30)  # 画左眼
t.end_fill()

# 画笑脸的右眼
t.penup()
t.goto(70, 50)  # 移动到右眼的位置
t.pendown()
t.begin_fill()
t.circle(30)  # 画右眼
t.end_fill()

# 画笑脸的嘴巴
t.penup()
t.goto(-100, -50)  # 移动到嘴巴的起点
t.setheading(-60)  # 设置海龟方向为-60度
t.pendown()
t.circle(120, 120)  # 画弧形，半径120，角度120度

# 完成绘画，隐藏海龟
t.hideturtle()

# 保持窗口打开
turtle.done()
```
