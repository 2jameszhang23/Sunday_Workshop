```python
import pygame
import time
import random

# 初始化 Pygame
pygame.init()

# 定义屏幕大小
width, height = 600, 400
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('贪吃蛇')

# 定义颜色
black = (0, 0, 0)
white = (255, 255, 255)
red = (213, 50, 80)
green = (0, 255, 0)
blue = (50, 153, 213)

# 定义蛇的块大小和速度
snake_block = 10
snake_speed = 15

# 设置字体
font_style = pygame.font.SysFont("bahnschrift", 25)

# 读取排行榜文件
def read_leaderboard():
    try:
        with open("leaderboard.txt", "r") as f:
            scores = [line.strip().split(":") for line in f.readlines()]
            scores = [(name, int(score)) for name, score in scores]
            return sorted(scores, key=lambda x: x[1], reverse=True)
    except FileNotFoundError:
        return []

# 保存排行榜
def save_leaderboard(name, score):
    scores = read_leaderboard()
    scores.append((name, score))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    with open("leaderboard.txt", "w") as f:
        for name, score in scores[:5]:  # 仅保存前5名
            f.write(f"{name}:{score}\n")

# 显示排行榜
def show_leaderboard():
    scores = read_leaderboard()
    screen.fill(black)
    message("Leaderboard:", white, (width / 6, height / 6))
    for i, (name, score) in enumerate(scores[:5]):
        message(f"{i+1}. {name}: {score}", white, (width / 6, height / 6 + (i + 1) * 30))
    pygame.display.update()
    time.sleep(5)

# 计分函数
def score(score):
    value = font_style.render("Your Score: " + str(score), True, white)
    screen.blit(value, [0, 0])

# 绘制蛇函数
def draw_snake(snake_block, snake_list):
    for x in snake_list:
        pygame.draw.rect(screen, green, [x[0], x[1], snake_block, snake_block])

# 显示消息函数
def message(msg, color, position):
    mesg = font_style.render(msg, True, color)
    screen.blit(mesg, position)

# 游戏循环
def gameLoop():
    game_over = False
    game_close = False

    # 初始蛇的位置和长度
    x = width / 2
    y = height / 2
    x_change = 0
    y_change = 0
    snake_list = []
    length_of_snake = 1

    # 食物的位置
    food_x = round(random.randrange(0, width - snake_block) / 10.0) * 10.0
    food_y = round(random.randrange(0, height - snake_block) / 10.0) * 10.0

    clock = pygame.time.Clock()

    while not game_over:

        while game_close == True:
            screen.fill(black)
            message("You Lost! Press C to Play Again or Q to Quit", red, (width / 6, height / 3))
            score(length_of_snake - 1)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        gameLoop()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x_change = -snake_block
                    y_change = 0
                elif event.key == pygame.K_RIGHT:
                    x_change = snake_block
                    y_change = 0
                elif event.key == pygame.K_UP:
                    y_change = -snake_block
                    x_change = 0
                elif event.key == pygame.K_DOWN:
                    y_change = snake_block
                    x_change = 0

        # 如果蛇超出边界，游戏结束
        if x >= width or x < 0 or y >= height or y < 0:
            game_close = True

        x += x_change
        y += y_change
        screen.fill(black)

        # 画食物
        pygame.draw.rect(screen, white, [food_x, food_y, snake_block, snake_block])

        # 更新蛇的身体
        snake_head = []
        snake_head.append(x)
        snake_head.append(y)
        snake_list.append(snake_head)

        if len(snake_list) > length_of_snake:
            del snake_list[0]

        for segment in snake_list[:-1]:
            if segment == snake_head:
                game_close = True

        draw_snake(snake_block, snake_list)
        score(length_of_snake - 1)

        pygame.display.update()

        # 吃到食物
        if x == food_x and y == food_y:
            food_x = round(random.randrange(0, width - snake_block) / 10.0) * 10.0
            food_y = round(random.randrange(0, height - snake_block) / 10.0) * 10.0
            length_of_snake += 1

        clock.tick(snake_speed)

    # 游戏结束，保存得分
    player_name = input("Enter your name: ")
    save_leaderboard(player_name, length_of_snake - 1)
    show_leaderboard()
    pygame.quit()
    quit()

# 启动游戏
gameLoop()
```
